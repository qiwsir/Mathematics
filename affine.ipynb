```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
  
  
img = cv2.imread('headpic.png') 
rows, cols, ch = img.shape 
  
pts1 = np.float32([[500, 500], 
                   [2000, 1000],  
                   [1500, 2000]]) 
  
pts2 = np.float32([[500, 1000], 
                   [1500, 1000],  
                   [1500, 2000]]) 
  
M = cv2.getAffineTransform(pts1, pts2) 
dst = cv2.warpAffine(img, M, (cols, rows)) 
  
plt.subplot(121) 
plt.imshow(img) 
plt.title('Input') 
  
plt.subplot(122) 
plt.imshow(dst) 
plt.title('Output') 
```

```python
# 平移
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("headpic.png")
T = np.float32([[1,0,500],[0,1,1000]])
rows, cols, ch = img.shape

res = cv2.warpAffine(img,T,(rows,cols)) #需要图像、变换矩阵、变换后的大小

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(res)
plt.title('Output') 
```

```python
C = np.float32([[0.5,0,0],[0,0.5,0]])

res = cv2.warpAffine(img,C,(rows//2,cols//2)) 

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(res)
plt.title('Output') 
```

```python
# res1 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

res2 = cv2.resize(img,(rows//2,cols//2))

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

# plt.subplot(132)
# plt.imshow(res1)
# plt.title('Output') 

plt.subplot(122)
plt.imshow(res2)
plt.title('Output')
```

```python
M = np.float32([[0.702, -0.702, 0],[0.702, 0.702, 0]])

res = cv2.warpAffine(img,M,(4000,4000)) 

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(res)
plt.title('Output') 
```

```python
M = cv2.getRotationMatrix2D((rows/2, cols/2), 45, 1)
res = cv2.warpAffine(img,M,(rows,cols)) 

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(res)
plt.title('Output') 
```
